# 호주 이민 포인트 시뮬레이션

호주 이민청(DHA) 포인트 테스트 기반 문제들

65점이 커트라인, 셈글 마스터는 90점대 노린다 ㅋㅋㅋ

## 문제 1: 나이 점수 계산

나이별 점수:
- 18-24: 25점
- 25-32: 30점
- 33-39: 25점
- 40-44: 15점
- 45+: 0점

나이를 입력받아 점수를 반환하는 함수를 만들어라.

```python
def calc_age_score(age):
    if 18 <= age <= 24:
        return 25
    elif 25 <= age <= 32:
        return 30
    elif 33 <= age <= 39:
        return 25
    elif 40 <= age <= 44:
        return 15
    else:
        return 0

# 테스트
print(calc_age_score(17), calc_age_score(28), calc_age_score(50))
```

## 문제 2: 영어 점수 계산

IELTS 점수별:
- 6.0 이하: 0
- 6.5: 10점
- 7.0: 15점
- 8.0 이상: 20점

함수 calc_english_score(ielts)를 만들어라.

```python
def calc_english_score(ielts):
    if ielts >= 8.0:
        return 20
    elif ielts >= 7.0:
        return 15
    elif ielts >= 6.5:
        return 10
    else:
        return 0

print(calc_english_score(5.5), calc_english_score(7.5))
```

## 문제 3: 학력 점수 계산

학력별:
- 고졸: 10점
- 기능고/전문대: 15점
- 대졸/석사/박사: 20점

함수 calc_edu_score(edu)를 만들어라.

```python
def calc_edu_score(edu):
    if edu in ['고졸']:
        return 10
    elif edu in ['기능고', '전문대']:
        return 15
    else:
        return 20

print(calc_edu_score('기능고'), calc_edu_score('대졸'))
```

## 문제 4: 경력 점수 계산

경력(개월)별:
- 12개월 미만: 0
- 12-35개월: 5점
- 36-59개월: 10점
- 60-95개월: 15점
- 96개월+: 20점

함수 calc_exp_score(months)를 만들어라.

```python
def calc_exp_score(months):
    if months >= 96:
        return 20
    elif months >= 60:
        return 15
    elif months >= 36:
        return 10
    elif months >= 12:
        return 5
    else:
        return 0

print(calc_exp_score(6), calc_exp_score(24), calc_exp_score(100))
```

## 문제 5: 종합 계산

위 함수들을 사용해서 총점을 계산하고 65점 이상이면 이민가능 출력

```python
age = int(input('나이: '))
ielts = float(input('IELTS: '))
edu = input('학력: ')
months = int(input('경력(개월): '))

total = calc_age_score(age) + calc_english_score(ielts) + calc_edu_score(edu) + calc_exp_score(months)
print(f'총점: {total}')
if total >= 65:
    print('이민 가능!')
else:
    print('커트라인 미달...')
```

## 문제 6: 가산점 계산

예/아니오 6개 항목 입력받아 총 가산점 계산 (최대 15점)

- 호주학위?
- 호주경력?
- 파트너영어?
- PY수료?
- NAATI?
- 지방거주?

```python
items = ['호주학위', '호주경력', '파트너영어', 'PY수료', 'NAATI', '지방거주']
count = 0
for item in items:
    ans = input(f'{item}? (예/아니오): ')
    if ans == '예':
        count += 1

bonus = min(count * 5, 15)
print(f'가산점: {bonus}')
```

## 문제 7: 셈글 마스터 프로필

셈글 마스터(호세) 정보로 점수 계산:
- 나이: 17세
- IELTS: 7.5
- 학력: 기능고
- 경력: 12개월

```python
master = {'age': 17, 'ielts': 7.5, 'edu': '기능고', 'months': 12}

score = calc_age_score(master['age']) + calc_english_score(master['ielts']) + calc_edu_score(master['edu']) + calc_exp_score(master['months'])
print(f'셈글 마스터 현재 점수: {score}')
print('IELTS 7.5와 경력 더 쌓아야 65점 돌파 가능!')
```

## 문제 8: 다중 지원자 비교

N명의 이름 나이 경력 영어점수 입력받아 각각 점수 계산하고
이민 가능자(50점+) 명단 출력

```python
n = int(input('지원자 수: '))
qualified = []

for _ in range(n):
    data = input('이름 나이 경력(월) 영어: ').split()
    name, age, months, ielts = data[0], int(data[1]), int(data[2]), float(data[3])
    
    total = calc_age_score(age) + calc_english_score(ielts) + calc_exp_score(months)
    if total >= 50:
        qualified.append((name, total))

print(f'이민 가능자: {qualified}')
```

IELTS 6.0~9.0까지 0.5단위로 점수 변화 테이블 출력

## 문제 10: 5년 플랜 시뮬레이션

2026~2030년 셈글 마스터 성장 시뮬레이션

```python
years = [2026, 2027, 2028, 2029, 2030]
ages = [17, 18, 19, 20, 21]
months = [0, 6, 18, 30, 42]
ielts = [0, 0, 6.0, 6.5, 7.0]

for i, year in enumerate(years):
    age_score = calc_age_score(ages[i])
    eng_score = calc_english_score(ielts[i])
    exp_score = calc_exp_score(months[i])
    edu_score = 15  # 기능고
    total = age_score + eng_score + edu_score + exp_score
    print(f'{year}년({ages[i]}세): {total}점')
```

## 문제 11: 실시간 점수 트래커

명령어 입력으로 점수 업데이트하는 시스템

- 나이 [숫자]
- 영어 [점수]
- 확인
- 종료

```python
age, ielts, months = 17, 0, 0

while True:
    cmd = input('명령: ').split()
    if cmd[0] == '나이':
        age = int(cmd[1])
    elif cmd[0] == '영어':
        ielts = float(cmd[1])
    elif cmd[0] == '확인':
        total = calc_age_score(age) + calc_english_score(ielts) + calc_exp_score(months)
        print(f'현재: {total}점')
    elif cmd[0] == '종료':
        break
```

## 문제 12: 비자 유형별 판정

189/190/491 비자별 가산점 적용해서 가능여부 출력

- 189: 기본점수
- 190: +5점
- 491: +15점

```python
base = int(input('기본점수: '))

visas = {'189': base, '190': base + 5, '491': base + 15}

for visa, score in visas.items():
    result = '가능' if score >= 65 else '불가'
    print(f'{visa} 비자: {score}점 - {result}')
```

## 문제 14: 연봉 환산 계산

경력별 AUD 연봉 범위를 원화로 환산

- 0-2년: 60,000-75,000 AUD
- 3-5년: 75,000-95,000 AUD
- 6년+: 95,000-130,000 AUD

1 AUD = 900 KRW

```python
years = int(input('경력(년): '))
rates = [900, 950, 1000]  # 환율 범위

if years <= 2:
    aud_low, aud_high = 60000, 75000
elif years <= 5:
    aud_low, aud_high = 75000, 95000
else:
    aud_low, aud_high = 95000, 130000

for rate in rates:
    low = aud_low * rate
    high = aud_high * rate
    print(f'{rate}원 환율: {low:,}~{high:,}원')
```

## 문제 15: 비용 계산

이민 비용 항목별 계산

- 주신청자: 4,115 AUD
- 파트너: 2,055 AUD
- 자녀: 1,030 AUD/명
- 어학시험: 300 AUD
- 의료검진: 300 AUD
- 서류비: 200 AUD

```python
partner = input('파트너 있음? (예/아니오): ') == '예'
children = int(input('자녀 수: '))

base = 4115
if partner:
    base += 2055
base += children * 1030
base += 300 + 300 + 200  # 어학+의료+서류

print(f'총 비용: {base:,} AUD ({base * 900:,}원)')
```

## 문제 16: 최적 시점 계산

현재 나이 입력받아 25-32세 구간(30점)에 들어가려면 몇 년 후인지 계산

## 문제 17: 경력 최적화

현재 경력 개월 입력받고 목표 점수별 필요 개월 계산

- 5점(12개월)
- 10점(36개월)
- 15점(60개월)
- 20점(96개월)

```python
current = int(input('현재 경력(개월): '))
targets = [12, 36, 60, 96]

for target in targets:
    if current < target:
        need = target - current
        print(f'{target}개월까지 {need}개월 더 필요')
    else:
        print(f'{target}개월 - 이미 달성')
```

## 문제 18: IELTS 목표 역산

현재 총점(가산점 제외) 입력받고 65점 달성을 위해 필요한 최소 IELTS 계산

```python
current = int(input('현재 총점(영어제외): '))
needed = 65 - current

if needed <= 0:
    print('이미 65점 돌파!')
elif needed <= 10:
    print(f'IELTS 6.5 필요 (10점)')
elif needed <= 15:
    print(f'IELTS 7.0 필요 (15점)')
else:
    print(f'IELTS 8.0 필요 (20점) 혹은 다른 항목 보완')
```

## 문제 19: 실시간 득표율

후보 A,B,C에게 N명이 투표. 과반(50%+) 나오면 즉시 종료

```python
n = int(input('총 투표자: '))
a = b = c = 0

for i in range(n):
    vote = input(f'{i+1}번째 투표(A/B/C): ').upper()
    if vote == 'A':
        a += 1
    elif vote == 'B':
        b += 1
    else:
        c += 1
    
    total = a + b + c
    print(f'A:{a/total*100:.1f}% B:{b/total*100:.1f}% C:{c/total*100:.1f}%')
    
    if a/total > 0.5 or b/total > 0.5 or c/total > 0.5:
        print('과반 달성! 종료!')
        break
```

모든 정보 입력받아 예쁘게 포맷팅해서 보고서 생성
