## 파트 1: 기본 함수 정답

```python
# 문제 1: 나이 점수
def calc_age_score(age):
    if 18 <= age <= 24:
        return 25
    elif 25 <= age <= 32:
        return 30
    elif 33 <= age <= 39:
        return 25
    elif 40 <= age <= 44:
        return 15
    else:
        return 0

# 테스트
print(calc_age_score(19), calc_age_score(28), calc_age_score(50))
```

```python
# 문제 2: 영어 점수
def calc_english_score(ielts):
    if ielts >= 8.0:
        return 20
    elif ielts >= 7.0:
        return 15
    elif ielts >= 6.5:
        return 10
    else:
        return 0

print(calc_english_score(5.5), calc_english_score(7.5))
```

```python
# 문제 3: 학력 점수
def calc_edu_score(edu):
    if edu == '고졸':
        return 10
    elif edu in ['기능고', '전문대']:
        return 15
    else:
        return 20

print(calc_edu_score('기능고'), calc_edu_score('대졸'))
```

```python
# 문제 4: 경력 점수
def calc_exp_score(months):
    if months >= 96:
        return 20
    elif months >= 60:
        return 15
    elif months >= 36:
        return 10
    elif months >= 12:
        return 5
    else:
        return 0

print(calc_exp_score(6), calc_exp_score(24), calc_exp_score(100))
```

```python
# 문제 5: 종합 계산
age = int(input('나이: '))
ielts = float(input('IELTS: '))
edu = input('학력: ')
months = int(input('경력(개월): '))

total = calc_age_score(age) + calc_english_score(ielts) + calc_edu_score(edu) + calc_exp_score(months)
print(f'총점: {total}')
print('이민가능' if total >= 65 else '미달')
```

## 파트 2: 응용 문제 정답

```python
# 문제 6: 가산점
items = ['호주학위', '호주경력', '파트너영어', 'PY수료', 'NAATI', '지방거주']
count = sum(1 for item in items if input(f'{item}? ') == '예')
bonus = min(count * 5, 15)
print(f'가산점: {bonus}')
```

```python
# 문제 7: 셈글 마스터
master = {'age': 17, 'ielts': 7.5, 'edu': '기능고', 'months': 12}

score = calc_age_score(master['age']) + calc_english_score(master['ielts']) + calc_edu_score(master['edu']) + calc_exp_score(master['months'])
print(f'현재 점수: {score}')
print(f'부족: {65-score}점')
print('-> 나이 25세까지 기다리거나 경력 더 쌓아야 함')
```

```python
# 문제 8: 다중 지원자
n = int(input())
qualified = []

for _ in range(n):
    data = input().split()
    name, age, months, eng = data[0], int(data[1]), int(data[2]), int(data[3])
    score = calc_age_score(age) + calc_exp_score(months) + (20 if eng >= 80 else 0)
    if score >= 50:
        qualified.append((name, score))

print(f'이민가능: {qualified}')
```

```python
# 문제 9: IELTS 테이블
scores = [6.0, 6.5, 7.0, 7.5, 8.0, 8.5, 9.0]

for ielts in scores:
    eng = calc_english_score(ielts)
    total = 30 + eng + 20 + 10  # 나이+영어+학력+경력
    print(f'IELTS {ielts}: {total}점')
```

```python
# 문제 10: 5년 플랜
years = [2026, 2027, 2028, 2029, 2030]
ages = [17, 18, 19, 20, 21]
months = [0, 6, 18, 30, 42]
ielts_target = [0, 0, 6.0, 6.5, 7.0]

for i, year in enumerate(years):
    age_score = calc_age_score(ages[i])
    eng_score = calc_english_score(ielts_target[i])
    exp_score = calc_exp_score(months[i])
    total = age_score + eng_score + 15 + exp_score  # 학력 15점(기능고)
    status = '커트라인돌파!' if total >= 65 else '더공부필요'
    print(f'{year}년({ages[i]}세): {total}점 - {status}')
```

```python
# 문제 11: 실시간 트래커
age, ielts, months = 17, 0, 0

while True:
    cmd = input('명령: ').split()
    if cmd[0] == '나이':
        age = int(cmd[1])
    elif cmd[0] == '영어':
        ielts = float(cmd[1])
    elif cmd[0] == '확인':
        total = calc_age_score(age) + calc_english_score(ielts) + 15 + calc_exp_score(months)
        print(f'현재: {total}점')
    elif cmd[0] == '종료':
        break
```

```python
# 문제 12: 비자별 판정
base = int(input('기본점수: '))

visas = {'189': base, '190': base + 5, '491': base + 15}

for visa, score in visas.items():
    result = '가능' if score >= 65 else '불가'
    print(f'{visa}: {score}점 ({result})')
```

```python
# 문제 13: 성공 확률
score = int(input('점수: '))

probabilities = {
    90: '95% - 금상첨화',
    80: '80% - 충분',
    70: '60% - 가능',
    65: '40% - 경쟁 필요',
    60: '20% - 어려움'
}

for threshold, msg in sorted(probabilities.items(), reverse=True):
    if score >= threshold:
        print(msg)
        break
else:
    print('5% - 불가')
```

```python
# 문제 14: 연봉 환산
years = int(input('경력(년): '))

if years <= 2:
    low, high = 60000, 75000
elif years <= 5:
    low, high = 75000, 95000
else:
    low, high = 95000, 130000

print(f'{low*900:,}~{high*900:,}원')
```

```python
# 문제 15: 비용 계산
partner = input('파트너? (예/아니오): ') == '예'
children = int(input('자녀 수: '))

base = 4115
if partner:
    base += 2055
base += children * 1030 + 800  # 기타비용

print(f'총 {base:,} AUD ({base*900:,}원)')
```

```python
# 문제 16: 최적 시점
age = int(input('현재 나이: '))

if age < 25:
    wait = 25 - age
    print(f'{wait}년 후 25세 되면 30점 만점!')
elif age <= 32:
    print('지금이 최적 타이밍!')
else:
    print('이미 30점 구간 지남...')
```

```python
# 문제 17: 경력 최적화
current = int(input('현재 경력(개월): '))
targets = [(12, 5), (36, 10), (60, 15), (96, 20)]

for target_months, score in targets:
    if current < target_months:
        need = target_months - current
        print(f'{score}점까지 {need}개월 더 필요')
    else:
        print(f'{score}점 - 달성')
```

```python
# 문제 18: IELTS 역산
current = int(input('현재 총점(영어제외): '))
need = 65 - current

if need <= 0:
    print('이미 돌파!')
elif need <= 10:
    print('IELTS 6.5 필요')
elif need <= 15:
    print('IELTS 7.0 필요')
else:
    print('IELTS 8.0 필요 or 다른 항목 보완')
```

```python
# 문제 19: 실시간 선거
n = int(input('총 투표자: '))
a = b = c = 0

for i in range(n):
    vote = input(f'{i+1}번째: ').upper()
    if vote == 'A': a += 1
    elif vote == 'B': b += 1
    else: c += 1
    
    total = a + b + c
    print(f'A:{a/total*100:.1f}% B:{b/total*100:.1f}% C:{c/total*100:.1f}%')
    
    if max(a, b, c) / total > 0.5:
        print('과반 달성!')
        break
```

```python
# 문제 20: FINAL 보고서
name = input('이름: ')
age = int(input('나이: '))
ielts = float(input('IELTS: '))
edu = input('학력: ')
months = int(input('경력(개월): '))

age_s = calc_age_score(age)
eng_s = calc_english_score(ielts)
edu_s = calc_edu_score(edu)
exp_s = calc_exp_score(months)
total = age_s + eng_s + edu_s + exp_s

print('='*40)
print('    호주 이민 준비 완료 보고서')
print('='*40)
print(f'성명: {name}')
print(f'현재 {age}세')
print('-'*40)
print(f'나이: {age_s}점 | 영어: {eng_s}점')
print(f'학력: {edu_s}점 | 경력: {exp_s}점')
print('-'*40)
print(f'총점: {total}점')
print(f'판정: {"가능!" if total >= 65 else "미달..."}')
print('='*40)
```
