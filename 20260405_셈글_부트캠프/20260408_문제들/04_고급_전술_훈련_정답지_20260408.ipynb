## for문 기초 정답

```python
# Q1. 구구단
n = int(input())
for i in range(1, 10):
    print(f'{n} * {i} = {n * i}')
```

```python
# Q2. A+B - 3
t = int(input())
for _ in range(t):
    a, b = map(int, input().split())
    print(a + b)
```

```python
# Q3. 합
n = int(input())
total = 0
for i in range(1, n + 1):
    total += i
print(total)
```

```python
# Q4. 영수증
x = int(input())
n = int(input())
total = 0
for _ in range(n):
    price, qty = map(int, input().split())
    total += price * qty
print('Yes' if total == x else 'No')
```

```python
# Q10. A+B - 5
while True:
    a, b = map(int, input().split())
    if a == 0 and b == 0:
        break
    print(a + b)
```

```python
# Q11. A+B - 4
while True:
    try:
        a, b = map(int, input().split())
        print(a + b)
    except EOFError:
        break
```

```python
# Q12. 더하기 사이클
n = int(input())
original = n
count = 0
while True:
    tens = n // 10
    ones = n % 10
    new_num = ones * 10 + (tens + ones) % 10
    count += 1
    if new_num == original:
        break
    n = new_num
print(count)
```

## 셈글 실전 시나리오 정답

```python
# Q15. 바리스타 연속 주문
prices = {'아메리카노': 2000, '라떼': 3000, '에스프레소': 1500}
n = int(input())
total = 0
for _ in range(n):
    menu, qty = input().split()
    qty = int(qty)
    total += prices[menu] * qty
print(f'총 매출: {total}원')
```

```python
# Q16. 원전 출력 모니터링
rates = list(map(int, input().split()))  # 24개

low_count = sum(1 for r in rates if r < 90)
print(f'90% 미만: {low_count}시간')

# 연속 최대 구간
max_streak = current = 0
for r in rates:
    if r < 90:
        current += 1
        max_streak = max(max_streak, current)
    else:
        current = 0
print(f'최대 연속 저하: {max_streak}시간')
```

```python
# Q17. 요트 항해 로그
n = int(input())
distances = []
for _ in range(n):
    distances.append(int(input()))

print(f'평균: {sum(distances)/n:.1f}km')
print(f'최대: {max(distances)}km ({distances.index(max(distances))+1}일째)')
print(f'총계: {sum(distances)}km')
```

```python
# Q18. 투표 집계 V2
n = int(input())
yes = no = 0
for _ in range(n):
    vote = input()
    if vote == '찬성':
        yes += 1
    else:
        no += 1

if yes > no:
    print(f'통과 (차이: {yes-no}표)')
elif no > yes:
    print(f'부결 (차이: {no-yes}표)')
else:
    print('재투표')
```

```python
# Q19. 재고 자동주문
inventory = {}
while True:
    cmd = input()
    if cmd == '주문':
        break
    bean, amount = cmd.split()
    inventory[bean] = int(amount)

for bean, amount in inventory.items():
    if amount < 50:
        print(f'{bean}: 재고부족({amount}kg)')
```

```python
# Q20. 공산주의 단어 감지
banned = ['공산주의', '탈원전']
alerts = 0
while True:
    text = input()
    if text == '종료':
        break
    for word in banned:
        if word in text:
            print('ALERT!!!')
            alerts += 1
            break
print(f'총 알람: {alerts}회')
```

## 복합 시나리오 정답

```python
# Q21. 이민 포인트 V2
def calc_score(age, months, english):
    score = 0
    if 18 <= age <= 30:
        score += 30
    if months >= 12:
        score += 10
    if english >= 80:
        score += 20
    return score

n = int(input())
count = 0
for _ in range(n):
    name, age, months, eng = input().split()
    score = calc_score(int(age), int(months), int(eng))
    if score >= 50:
        count += 1
print(f'이민 가능자: {count}명')
```

```python
# Q22. 원전 월간 보고서
outputs = [int(input()) for _ in range(30)]

print(f'총 발전량: {sum(outputs)}MWh')
print(f'일 평균: {sum(outputs)/30:.1f}MWh')
print(f'최대: {max(outputs)}MWh ({outputs.index(max(outputs))+1}일)')

low_days = [i+1 for i, o in enumerate(outputs) if o < 1000]
print(f'1000미만 날짜: {low_days}')
```

```python
# Q23. 주문 패턴
records = [input() for _ in range(7)]

# 최다 메뉴
from collections import Counter
most = Counter(records).most_common(1)[0][0]
print(f'최다 판매: {most}')

# 최대 연속
max_streak = curr = 1
for i in range(1, 7):
    if records[i] == records[i-1]:
        curr += 1
        max_streak = max(max_streak, curr)
    else:
        curr = 1
print(f'최대 연속: {max_streak}회')
```

```python
# Q24. 풍속 경보
n = int(input())
winds = [int(input()) for _ in range(n)]

danger = warning = normal = 0
for w in winds:
    if w >= 20:
        print('위험')
        danger += 1
    elif w >= 15:
        print('주의')
        warning += 1
    else:
        print('정상')
        normal += 1

print(f'위험:{danger}h 주의:{warning}h 정상:{normal}h')
```

## FINAL BOSS 정답

```python
# Q26. 카페 매출 분석
sales = [int(input()) for _ in range(30)]

# 주간별 평균
weekly_avgs = []
for week in range(4):
    start = week * 7
    end = min(start + 7, 30)
    avg = sum(sales[start:end]) / (end - start)
    weekly_avgs.append(avg)

best_week = weekly_avgs.index(max(weekly_avgs)) + 1
print(f'최고 주간: {best_week}주차')

high_days = sum(1 for s in sales if s >= 50000)
print(f'5만원↑: {high_days}일')

decrease_days = sum(1 for i in range(1, 30) if sales[i] < sales[i-1])
print(f'전일대비 감소: {decrease_days}일')
```

```python
# Q27. 원전 비상 대피
rates = []
while True:
    cmd = input()
    if cmd == '종료':
        break
    rate = int(cmd)
    rates.append(rate)
    
    if rate < 50:
        print('비상! 비상! 셧다운 준비!')
        break
    elif rate < 70:
        print('주의: 출력 하락')
    else:
        print('정상')

if rates:
    print(f'평균 출력률: {sum(rates)/len(rates):.1f}%')
```

```python
# Q28. 요트 횡단
daily = int(input())
remaining = 2000
days = 0
bad_weather = False

while remaining > 0:
    days += 1
    if bad_weather:
        travel = daily * 0.5
    else:
        travel = daily
    remaining -= travel
    bad_weather = not bad_weather  # 바람 안좋은날 번갈아

print(f'{days}일 소요')
```

```python
# Q29. 고양이민주주의 선거
n = int(input())
a = b = c = 0

for i in range(n):
    vote = input().upper()
    if vote == 'A':
        a += 1
    elif vote == 'B':
        b += 1
    else:
        c += 1
    
    total = a + b + c
    print(f'A:{a/total*100:.1f}% B:{b/total*100:.1f}% C:{c/total*100:.1f}%')
    
    if a/total > 0.5 or b/total > 0.5 or c/total > 0.5:
        print('과반 달성!')
        break
```

```python
# Q30. 마스터의 인벤토리
inventory = {}

while True:
    cmd = input().split()
    action = cmd[0]
    
    if action == '종료':
        break
    elif action == '확인':
        for item, qty in inventory.items():
            print(f'{item}: {qty}')
    elif action == '추가':
        item, qty = cmd[1], int(cmd[2])
        inventory[item] = inventory.get(item, 0) + qty
    elif action == '삭제':
        item, qty = cmd[1], int(cmd[2])
        if item not in inventory:
            print('없음')
        elif inventory[item] < qty:
            print('부족함')
        else:
            inventory[item] -= qty
```
