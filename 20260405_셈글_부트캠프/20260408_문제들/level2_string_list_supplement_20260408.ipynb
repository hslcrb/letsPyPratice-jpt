### Q3. 숫자의 합 (BOJ 11720)

### Q5. 문자열 반복 (BOJ 2675)

각 문자를 R번 반복해서 새 문자열

```python
t = int(input())
for _ in range(t):
    r, s = input().split()
    r = int(r)
    result = ''
    for c in s:
        result += c * r
    print(result)
```

### Q7. 상수 (BOJ 2908)

숫자 두 개 뒤집어서 큰 수 출력. [::-1] 슬라이싱

```python
a, b = input().split()
a_rev = int(a[::-1])
b_rev = int(b[::-1])
print(max(a_rev, b_rev))
```

### Q8. 다이얼 (BOJ 5622)

전화번호 걸리는 시간 계산. ABC=3초, DEF=4초...

```python
s = input()
time = 0
for c in s:
    if c in 'ABC':
        time += 3
    elif c in 'DEF':
        time += 4
    elif c in 'GHI':
        time += 5
    elif c in 'JKL':
        time += 6
    elif c in 'MNO':
        time += 7
    elif c in 'PQRS':
        time += 8
    elif c in 'TUV':
        time += 9
    else:
        time += 10
print(time)
```

### Q9. 크로아티아 알파벳 (BOJ 2941)

c=, dz=, d-, lj, nj, s=, z=을 한 글자로 치고 총 글자 수 세기

```python
s = input()
croatia = ['c=', 'c-', 'dz=', 'd-', 'lj', 'nj', 's=', 'z=']
for c in croatia:
    s = s.replace(c, '*')
print(len(s))
```

### Q10. 그룹 단어 체커 (BOJ 1316)

같은 글자가 연속해서 나오는 단어 개수 세기

```python
n = int(input())
count = 0
for _ in range(n):
    word = input()
    seen = []
    is_group = True
    for i, c in enumerate(word):
        if c in seen and c != word[i-1]:
            is_group = False
            break
        if c not in seen:
            seen.append(c)
    if is_group:
        count += 1
print(count)
```

### Q15. 공 넣기 (BOJ 10810)

N개 바구니에 M번 공 넣기. 슬라이싱 활용

```python
n, m = map(int, input().split())
baskets = [0] * (n + 1)

for _ in range(m):
    i, j, k = map(int, input().split())
    for idx in range(i, j + 1):
        baskets[idx] = k

for i in range(1, n + 1):
    print(baskets[i], end=' ')
```

### Q16. 공 바꾸기 (BOJ 10813)

M번 두 바구니 공 교환. 다중 할당 활용

```python
n, m = map(int, input().split())
baskets = list(range(n + 1))

for _ in range(m):
    i, j = map(int, input().split())
    baskets[i], baskets[j] = baskets[j], baskets[i]

for i in range(1, n + 1):
    print(baskets[i], end=' ')
```

### Q17. 과제 안 낸 사람 (BOJ 5597)

30명 중 28명 출석번호 주어질 때 안 낸 2명 출력

```python
all_nums = set(range(1, 31))
submitted = set()
for _ in range(28):
    submitted.add(int(input()))

missing = sorted(all_nums - submitted)
for num in missing:
    print(num)
```

### Q18. 나머지 (BOJ 3052)

10개 수를 42로 나눈 나머지 중 서로 다른 값 개수

```python
remainders = set()
for _ in range(10):
    num = int(input())
    remainders.add(num % 42)
print(len(remainders))
```

### Q19. 바구니 뒤집기 (BOJ 10811)

바구니 범위를 뒤집는다

```python
n, m = map(int, input().split())
baskets = list(range(n + 1))

for _ in range(m):
    i, j = map(int, input().split())
    baskets[i:j+1] = baskets[i:j+1][::-1]

for i in range(1, n + 1):
    print(baskets[i], end=' ')
```

### Q20. 평균 (BOJ 1546)

최고점 M으로 모든 점수를 M/100*점수 로 계산한 평균

```python
n = int(input())
scores = list(map(int, input().split()))
m = max(scores)

new_scores = [(s / m) * 100 for s in scores]
print(sum(new_scores) / n)
```

### Q21. 바리스타 메모 파싱

메모 아메리카노3,에스프레소2,라떼1 에서 총 잔수 구하기

```python
memo = '아메리카노3,에스프레소2,라떼1'
items = memo.split(',')
total = 0
for item in items:
    count = int(item[-1])  # 마지막 문자가 숫자
    total += count
print(f'총 {total}잔')
```

### Q22. 고양이민주주의 투표 집계

찬성/반대 리스트에서 다수결 확인

```python
votes = ['찬성', '반대', '찬성', '찬성', '반대', '찬성']
yes = votes.count('찬성')
no = votes.count('반대')

if yes > no:
    print(f'고양이민주주의 통과! (차이: {yes-no}표)')
else:
    print(f'선동 detected... (차이: {no-yes}표)')
```

### Q23. 원전 로그 분석

10일간 원전 출력률 평균 90% 미만인 날 수 세기

```python
rates = [92, 88, 95, 91, 89, 94, 90, 87, 93, 96]
low_count = 0
for rate in rates:
    if rate < 90:
        low_count += 1

avg = sum(rates) / len(rates)
print(f'평균: {avg:.1f}%')
print(f'90% 미만: {low_count}일')
```

### Q25. 마스터의 비밀 메시지

### Q26. 커피 원두 재고 관리

재고 70kg 이하인 원두 인덱스 출력

```python
stocks = [100, 80, 95, 60, 70]  # 아라비카, 로부스타, 콜롬비아, 에티오피아, 케냐
names = ['아라비카', '로부스타', '콜롬비아', '에티오피아', '케냐']

for i, stock in enumerate(stocks):
    if stock <= 70:
        print(f'{names[i]}: 재고부족({stock}kg)')
```

### Q27. 금지단어 필터링

문장에 금지단어가 있으면 검열처리

```python
banned = ['공산주의', '탈원전', '감성팔이', '위선']
text = input('문장: ')

found = False
for word in banned:
    if word in text:
        found = True
        break

print('검열처리' if found else '정상')
```

### Q28. 학생 점수 통계

5명 점수 공백 구분 입력받아 평균, 최고, 최저

### Q30. FINAL: 마스터의 수능 영어

a로 시작하는 단어만 골라 대문자로 변환 후 정렬

```python
words = ['abandon', 'ability', 'abolish', 'absolute', 'abstract', 'zebra', 'apple']

# a로 시작하는 단어 필터링
a_words = [w for w in words if w.startswith('a')]
# 대문자 변환
a_words_upper = [w.upper() for w in a_words]
# 정렬
a_words_upper.sort()

print(a_words_upper)
```
