# 🎯 셈글 파이썬 실전 참교육 문제 (마스터 정답지)

호세 마스터의 자본주의와 팩트 폭격이 담긴 실전 문제다 ㅋㅋㅋ 이건 정답지니까 절대 후배들에게 유출하지 마라!

## ☕ 문제 1: 바리스타 호세의 철저한 자본주의 계산기

진상 손님이 와서 에스프레소(1500원) 샷 개수와 아메리카노(2000원) 잔 수를 공백을 두고 한 번에 입력한다.
빈틈없는 자본주의의 맛으로 결제 금액을 계산해서 출력해라.

```python
e_count, a_count = map(int, input("에스프레소 잔 수와 아아 잔 수를 띄어쓰기로 입력해라: ").split())
total_price = (e_count * 1500) + (a_count * 2000)

print("어이 진상 손님, 에스프레소", e_count, "샷이랑 아아", a_count, "잔 해서 총", total_price, "원이다!")
print("공짜 배급은 없으니 빨리 결제해라 ㅋㅋㅋ")
```

## ⛵ 문제 2: 호주 크리스탈 워터스로 가는 1인승 요트

1인승 요트를 타고 호주로 가는 중 장애물을 만났다!
풍속(wind)과 파도 높이(wave)를 한 번에 띄어쓰기로 입력받아 생사를 팩트체크해라.

- 풍속 > 20 이거나 파도 > 5 이면: `가짜뉴스급 태풍이다! 팩트에 기반해 당장 돛을 접어라 으아아!!`
- 풍속 > 10 이고 파도 > 3 이면: `위선적인 감성팔이 역풍이다! 긴장 타라!`
- 그 외: `평화로운 고양이민주주의 바다... 직진하자 ㅋㅋㅋ!`

```python
wind, wave = map(int, input("풍속과 파도의 높이를 띄어쓰기로 입력해라: ").split())

if wind > 20 or wave > 5:
    print("가짜뉴스급 태풍이다! 팩트에 기반해 당장 돛을 접어라 으아아!!")
elif wind > 10 and wave > 3:
    print("위선적인 감성팔이 역풍이다! 긴장 타라!")
else:
    print("평화로운 고양이민주주의 바다... 직진하자 ㅋㅋㅋ!")
```

## 🐈 문제 3: 고양이민주주의 첩자 색출기

문장을 통째로 하나 입력받아서 띄어쓰기(`.split()`) 기준으로 쪼개라 (리스트 생성).
쪼갠 단어 리스트 안에 "탈원전" 또는 "공산주의" 가 포함되어 있으면 선동가를 검거하고, 아니면 훌륭한 시민으로 칭찬해라.

```python
sentence = input("조사할 문장을 통째로 입력해라: ")
word_list = sentence.split()

if "탈원전" in word_list or "공산주의" in word_list:
    print("삐빅- 위선적인 감성팔이 선동가 검거 완료 ㅋㅋㅋ! 똑바로 살아라!")
else:
    print("참된 본질적 우파 시민이구만. 환영한다!")
```

## ☢️ 문제 4 (구 FINAL BOSS): 원자력 발전소 출력 제어 시스템

실질적 에너지 안보를 위해 원자력 발전소의 출력을 관리하자 ㅋㅋㅋ
입력: `상태(str)`, `현재출력(int)`, `목표출력(int)`을 공백으로 구분해서 입력받아라.

- 만약 상태가 `"점검"` 이라면: `"위선적인 가짜뉴스로 인한 셧다운 중지! 팩트 체크 후 재가동 하라!"` 출력
- 만약 상태가 `"가동"` 이라면:
    - 현재출력 < 목표출력: `"에너지 안보를 위해 출력을 더 높여라 ㅋㅋㅋ!"` 출력
    - 현재출력 > 목표출력: `"과부하 위험! 감성 다 빼고 냉철하게 출력을 낮춰라!"` 출력
    - 현재출력 == 목표출력: `"안정적인 에너지 공급 중. 고양이민주주의의 번영이 눈앞에 있다 ㅋㅋㅋ"` 출력

```python
ipt = input("상태(가동/점검) 현재출력 목표출력을 입력해라: ").split()
status = ipt[0]
current = int(ipt[1])
target = int(ipt[2])

if status == "점검":
    print("위선적인 가짜뉴스로 인한 셧다운 중지! 팩트 체크 후 재가동 하라!")
elif status == "가동":
    if current < target:
        print("에너지 안보를 위해 출력을 더 높여라 ㅋㅋㅋ!")
    elif current > target:
        print("과부하 위험! 감성 다 빼고 냉철하게 출력을 낮춰라!")
    else:
        print("안정적인 에너지 공급 중. 고양이민주주의의 번영이 눈앞에 있다 ㅋㅋㅋ")
else:
    print("유물론적 오류 발생! 상태를 똑바로 입력해라 ㅋㅋㅋ")
```

## 🇦🇺 문제 5: 호세 마스터의 호주 이민 점수 계산기

호세 마스터의 호주 이민 꿈을 응원하며 점수를 계산해보자! ㅋㅋㅋ
입력: `나이(int)`, `경력(개월, int)`, `영어점수(int)`

- 점수 규칙:
    - 나이가 18~30세이면 30점, 아니면 0점 (팩트임 ㅋㅋㅋ)
    - 경력이 12개월 이상이면 10점, 아니면 0점
    - 영어점수가 80점 이상이면 20점, 아니면 0점
- 결과: 총점이 50점 이상이면 `"이민 쌉가능! 호주에서 요트 타자 ㅋㅋㅋ"`, 아니면 `"셈글에서 파이썬이나 더 해라!"` 출력

```python
age, exp, eng = map(int, input("나이 경력 영어점수를 입력해라: ").split())
score = 0

if 18 <= age <= 30:
    score += 30
if exp >= 12:
    score += 10
if eng >= 80:
    score += 20

if score >= 50:
    print("총점", score, ": 이민 쌉가능! 호주에서 요트 타자 ㅋㅋㅋ")
else:
    print("총점", score, ": 셈글에서 파이썬이나 더 해라!")
```

## ☕ 문제 6: 커피숍 재고 알리미 (재택근무 차단용)

재택근무하면서 월급 루팡하는 놈들 차단하기 위해 원두 재고를 체크하자 ㅋㅋㅋ
입력: `원두무게(g)`, `물(ml)`, `주문잔수`
- 1잔당 원두 18g, 물 200ml 필요.
- 재고가 충분하면 `"정상 가동! 카페인 풀충전 ㅋㅋㅋ"` 출력
- 부족하면 `"재고 부족! 유물론적 한계에 부딪혔다!"` 출력

```python
beans, water, order = map(int, input("원두(g) 물(ml) 주문수: ").split())

if beans >= order * 18 and water >= order * 200:
    print("정상 가동! 카페인 풀충전 ㅋㅋㅋ")
else:
    print("재고 부족! 유물론적 한계에 부딪혔다!")
```

## ☢️ 문제 7: 원전 가동 효율(이용률) 판독기

환경 팔이들의 선동에 맞서 원전의 실제 효율을 계산해보자!
입력: `실제발전량`, `최대발전가능량`
- 이용률(%) = (실제 / 최대) * 100
- 이용률이 90% 이상이면 `"킹갓 원자력! 에너지 안보 든든하다 ㅋㅋㅋ"` 출력
- 아니면 `"점검 필요! 팩트 체크 들어간다!"` 출력

```python
actual, max_val = map(int, input("실제발전량 최대발전량 입력: ").split())
rate = (actual / max_val) * 100

print("가동 이용률:", rate, "%")
if rate >= 90:
    print("킹갓 원자력! 에너지 안보 든든하다 ㅋㅋㅋ")
else:
    print("점검 필요! 팩트 체크 들어간다!")
```

## 🏴‍☠️ FINAL BOSS: 대양의 좌표 파싱 (요트 항해의 정점)

1인승 요트로 망망대해를 건너려면 좌표 정보 파싱은 기본이다 ㅋㅋㅋ
입력: `경도,위도,속도` (콤마로 구분된 문자열)
- `split(',')`을 써서 각각 분리해라.
- 속도가 20노트 이상이면 `"쾌속 순항! 팩트 폭격 날씨구만!"` 출력
- 아니면 `"저속 주행... 장 건강이나 챙기자 ㅋㅋㅋ"` 출력

```python
data = input("경도,위도,속도 순으로 입력 (예: 127.5,37.2,25) : ").split(',')
lon = data[0]
lat = data[1]
speed = float(data[2])

print("현재 위치: 경도", lon, "위도", lat)
if speed >= 20:
    print("속도", speed, "노트: 쾌속 순항! 팩트 폭격 날씨구만!")
else:
    print("속도", speed, "노트: 저속 주행... 장 건강이나 챙기자 ㅋㅋㅋ")
```
