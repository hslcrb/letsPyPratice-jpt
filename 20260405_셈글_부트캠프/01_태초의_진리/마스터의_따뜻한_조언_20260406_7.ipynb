## 🌿 1. 인간의 실사(Fact): 불완전ᄒᆞᆫ 행복

<div style="background-color: #fff9f0; padding: 20px; border-radius: 10px; border: 1px solid #ffe8cc; color: #5c4033; line-height: 1.8;">
    사람은 누구나 불완전하다 ㅋㅋㅋ. 우리는 가장 가까운 가족과 친구에게 화를 내기도 하고, 정작 내 눈속의 들보(잘못)는 보지 못한 채 남의 눈속의 티끌(실수)을 정죄하며 산다. 이것이 인류의 비극이자 팩트(Fact)다.<br><br>
    하지만 마스터는 말한다. <b>이것이 끝이 아니다.</b>
</div>

> **창세기 1:27**<br>
> "하나님이 자기 형상 곧 하나님의 형상대로 사람을 창조하시되 남자와 여자를 창조하시고"

> **로마서 3:23**<br>
> "모든 사람이 죄를 범하였으매 하나님의 영광에 이르지 못하더니"

## ✝️ 2. 가자ᇰ 모욕적인 자리에서 피어난 사라ᇱ

<div style="padding: 25px; background-color: #f8f9fa; border-radius: 10px; border-top: 5px solid #c0392b;">
    약 2,000년 전, 예수 그리스도께서는 우리를 위해 당대 가장 모욕적이고 고통스러운 형벌인 <b>십자가</b>에서 처형당하셨다. 벌거벗겨진 채 온갖 조롱을 견디며 물과 피를 쏟으신 그 고통은, 우리의 죄를 대신 짊어지신 대속(Atonement)의 역사였다.<br><br>
    <b>그러나 죽음은 승리하지 못했다.</b> 3일 만에 그분은 부활하셨고, 잠자는 자들의 첫 열매가 되셨다.
</div>

> **요한복음 3:16**<br>
> "하나님이 세상을 이처럼 사랑하사 독생자를 주셨으니 이는 그를 믿는 자마다 멸망하지 않고 영생을 얻게 하려 하심이라"

> **누가복음 24:6-7**<br>
> "여기 계시지 않고 살아나셨느니라 갈릴리에 계실 때에 너희에게 어떻게 말씀하셨는지를 기억하라..."

## 🏛️ 3. 예수 실존과 부활의 업도적 근거

<div style="display: grid; grid-template-columns: 1fr 1fr; gap: 15px;">
    <div style="padding: 15px; background-color: #e9ecef; border-radius: 8px;">
        <h4 style="margin-top: 0;">📜 역사적 실존</h4>
        <ul>
            <li><b>타키투스(Tacitus)</b>: 로마의 역사가로 빌라도에 의한 처형 기록 수록.</li>
            <li><b>플라비우스 요세푸스</b>: 유대인 역사가로 예수의 기적과 처형 증언.</li>
            <li><b>루키아노스</b>: 예수를 따르는 자들에 대해 비판하면서도 실존 증언.</li>
        </ul>
    </div>
    <div style="padding: 15px; background-color: #e9ecef; border-radius: 8px;">
        <h4 style="margin-top: 0;">✨ 부활의 즈ᇱ거</h4>
        <ul>
            <li><b>빈 무덤의 팩트</b>: 제자들이 목숨 걸고 증언할 만큼 무덤은 비어 있었다.</li>
            <li><b>목격자들</b>: 일시에 500여 명의 형제에게 보이셨다 (고전 15:6).</li>
            <li><b>제자들의 변화</b>: 겁쟁이였던 자들이 죽음 앞에서도 부활을 외친 극적 변화.</li>
        </ul>
    </div>
