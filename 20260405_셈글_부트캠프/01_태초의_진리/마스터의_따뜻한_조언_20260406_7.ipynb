## 🪞 1장. 우리는 왜 이렇게 엉망인가

솔직히 말해보자.

나도 그렇고, 너도 그렇고, 우리 모두는 가끔 진짜 못난 사람이 된다.

엄마한테 괜히 짜증 냈다가 방에 들어와서 혼자 후회하고,  
친한 친구한테 상처 주는 말을 했다가 나중에 미안해서 침대에서 이불 킥 하고,  
남이 잘못하면 눈에 엄청 잘 보이는데 내가 잘못하면 합리화부터 한다.  

이게 한두 번이 아니다. 패턴이다.

<div style='background: #f8f9fa; border-radius: 12px; padding: 20px 25px; margin: 15px 0; color: #495057; line-height: 1.8;'>
  성경은 이걸 2,000년 전에 이미 정확하게 진단했다.<br><br>
  <blockquote style='border-left: 4px solid #6c757d; padding-left: 20px; margin: 10px 0; color: #343a40; font-style: italic;'>
    "모든 사람이 죄를 범하였으매 하나님의 영광에 이르지 못하더니" <br>
    <strong style='font-style: normal; color: #6c757d;'>— 로마서 3:23</strong>
  </blockquote>
  여기서 '죄'는 법적인 범죄만 말하는 게 아니다.<br>
  원어로는 <em>'과녁을 빗나가다(ἁμαρτία, hamartia)'</em>는 뜻이다.<br>
  즉, 우리가 <strong>마땅히 되어야 할 모습</strong>에서 벗어난 상태. 그게 우리 모두의 기본 설정값이라는 거다.
</div>

그러니까 너 혼자 유독 이상한 게 아니다.  
이게 인류의 default 상태다. 나도, 역대 모든 위인도 다 마찬가지였다.

## ✨ 2장. 그런데 놀라운 반전이 있다

<div style='background: linear-gradient(135deg, #667eea 0%, #764ba2 100%); border-radius: 15px; padding: 30px; margin: 10px 0; color: white;'>
  <h3 style='margin: 0 0 15px 0; color: #fff; font-size: 1.3em;'>📌 역대 가장 중요한 팩트</h3>
  <p style='margin: 0; line-height: 1.9; font-size: 1.05em;'>
    "하나님이 세상을 이처럼 사랑하사 독생자를 주셨으니<br>
    이는 그를 믿는 자마다 멸망하지 않고 영생을 얻게 하려 하심이라"<br>
    <span style='font-size: 0.85em; color: #d4afff;'>— 요한복음 3:16</span>
  </p>

잠깐, 이 구절을 너무 익숙하게 그냥 넘기는 경우가 많다.  
근데 뒤에 나오는 내용이 그 맥락을 제대로 알면 정말 다르게 읽힌다.

**십자가형**은 단순한 사형이 아니었다.

로마 제국에서 십자가형은 일부러 최대한 오래, 최대한 고통스럽게,  
그리고 **최대한 모욕적으로** 죽이기 위해 설계된 형벌이었다.  
옷을 다 벗긴 채, 공중 앞에서, 조롱받으며 죽어가는 방식.  
심지어 로마 시민권자에게는 집행조차 금지됐던 — 노예와 최하층민에게만 쓰던 방법이었다.

<div style='background: #fff3cd; border: 1px solid #ffc107; border-radius: 10px; padding: 20px 25px; margin: 15px 0; line-height: 1.8;'>
  역사학자 키케로(Cicero)는 로마 시민에게 십자가 처형을 언급하는 것조차 "부끄러운 일"이라 기록했다.  
  그 정도로 치욕적인 형벌이었다.
</div>

예수님은 그 자리에 자발적으로 가셨다.  
피해서 살 수도 있었다. 실제로 군중을 피해 숨으신 적도 있다.  
하지만 마지막엔 피하지 않으셨다.  

왜냐면, 목적이 분명했으니까.  
**우리를 위해서.**

## 🏛️ 3장. 이게 '종교 얘기'라고 넘기기 전에 — 팩트 체크

예수님의 실존과 부활에 대해 "그냥 믿음으로 받아들여야 하는 것"이라고 생각하는 사람이 많다.  
근데 사실 역사적 증거가 상당히 압도적이다. 냉정하게 한번 보자.

### 📜 비기독교 역사가들의 기록

| 기록자 | 신분 | 기록 내용 |
|--------|------|----------|
| **타키투스 (Tacitus)** | 로마 역사가 (기독교인 아님) | 로마 황제 네로 시대에 "그리스도인들"이 대대적으로 박해받았으며, 그 창시자 '크리스투스'가 본디오 빌라도에 의해 처형됐다고 기록 (Annales, 116 AD) |
| **플라비우스 요세푸스 (Josephus)** | 유대인 역사가 | 예수라 불리는 현자의 기적과 십자가 처형, 그리고 부활 이후 그를 따르는 무리에 대해 기록 (Antiquities, 93 AD) |
| **루키아노스 (Lucian)** | 그리스 풍자작가 (기독교인 아님) | 그리스도인들이 팔레스타인에서 십자가에 처형된 "현자"를 숭배한다고 조롱하며 기록 |

> 기독교가 싫었거나 중립이었던 사람들도 예수의 실존 자체는 부정하지 않았다.  
> 이건 연구한 역사학자들의 공통된 결론이기도 하다.

### ⬛ 빈 무덤 — 반박이 가장 어려운 증거

<div style='background: #f1f3f4; border-radius: 10px; padding: 20px 25px; line-height: 1.8; margin: 10px 0;'>
  예수님 처형 후 3일 뒤, 무덤은 비어 있었다.<br><br>
  만약 예수님이 부활하지 않았다면, 예수를 처형한 <strong>로마 당국이나 유대 종교 지도자들이</strong> 시신을 꺼내 보여주면 됐다.  그렇게 하면 기독교 운동은 그 자리에서 끝이었다.<br><br>
  하지만 그들은 그러지 않았다. 못 했다. 왜냐하면 무덤이 실제로 비어 있었기 때문이다.  
  대신 그들이 퍼뜨린 말이 "제자들이 시체를 훔쳐갔다"는 것이었다. (마태복음 28:13)

### 🔥 제자들의 변화 — 가장 강력한 증거

예수님 처형 직전, 제자들은 겁에 질려 도망갔다.  
베드로는 예수님을 세 번이나 모른다고 부인했다.  
그들은 숨어서 문에 빗장을 걸고 떨고 있었다.

근데 몇 주 뒤, 이 사람들이 예루살렘 한복판에서 공개적으로 "예수님이 부활했다"고 외치기 시작했다.  
그리고 그것 때문에 고문받고, 감옥에 가고, 처형됐는데도 끝까지 철회하지 않았다.

<div style='background: #d4edda; border: 1px solid #28a745; border-radius: 10px; padding: 18px 22px; margin: 10px 0; line-height: 1.8;'>

  죽음 앞에서는 본음이 나온다. 그들이 끝까지 부활을 증언하며 죽어간 건,<br>
  그들이 진짜로 부활하신 예수님을 "눈으로 봤기" 때문이다.

  내가 너를 지명하여 불렀나니 너는 내 것이라"</strong><br>
  <span style='color: #6c757d;'>이사야 43:1</span>
</blockquote>

"내 것이라"는 히브리어 원어로는 소유를 뜻하는 게 아니라,  
**"내가 너를 안다, 내가 너를 향한 계획이 있다"** 는 뜻에 가깝다.  

고1이든 고2든, 뭘 해야 할지 모르겠고,  
내가 한심하게 느껴지는 날이 있어도,  
너라는 존재는 이미 하나님이 이름을 불러 창조하신 사람이다.
