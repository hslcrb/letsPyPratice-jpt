### Q2. 나이 계산기
올해 연도(2026)에서 태어난 연도(2008)를 빼서 나이를 계산해라.

```python
this_year = 2026
birth_year = 2008
age = this_year - birth_year + 1  # 한국 나이 기준
print("내 나이는", age, "살이다 ㅋㅋㅋ")
```

### Q3. 초간단 계산기
두 숫자 10과 5를 더하고, 빼고, 곱하고, 나눠서 출력해라.

```python
a, b = 10, 5
print("더하기:", a + b)
print("빼기:", a - b)
print("곱하기:", a * b)
print("나누기:", a / b)
```

### Q4. 삼각형 면적 구하기
밑변 10, 높이 5인 삼각형의 면적을 구해라. (밑변 * 높이 / 2)

```python
base = 10
height = 5
area = base * height / 2
print("삼각형 면적:", area)
```

### Q6. 행운의 숫자 7?
숫자 하나를 입력받아, 그 숫자가 7이면 "럭키 세븐 ㅋㅋㅋ!"을 출력하고 아니면 "가짜 숫자다!"를 출력해라.

```python
num = int(input("숫자를 입력해라: "))
if num == 7:
    print("럭키 세븐 ㅋㅋㅋ!")
else:
    print("가짜 숫자다!")
```

### Q7. 홀짝 판별기
숫자를 입력받아 2로 나눈 나머지가 0이면 "짝수", 아니면 "홀수"라고 출력해라.

```python
num = int(input("홀짝 확인할 숫자: "))
if num % 2 == 0:
    print("짝수 (진리)")
else:
    print("홀수 (팩트)")
```

### Q8. 성적 판독기
점수를 입력받아 90점 이상이면 A, 80점 이상이면 B, 그 외엔 F를 출력해라.

```python
score = int(input("점수 입력해라: "))
if score >= 90:
    print("A (에이스)")
elif score >= 80:
    print("B (보통)")
else:
    print("F (가짜뉴스)")
```

### Q9. 마스터 호세 환영 인사
이름을 입력받아 만약 "호세" 이면 "어서오십시오 마스터 ㅋㅋㅋ" 를 출력하고 아니면 "안녕 후배!" 를 출력해라.

```python
name = input("이름이 뭐냐: ")
if name == "호세":
    print("어서오십시오 마스터 ㅋㅋㅋ")
else:
    print("안녕 후배!")
```

### Q10. BMI 맛보기
체중(kg)을 키(m)의 제곱으로 나눈다. 몸무게 70kg, 키 1.8m인 사람의 BMI를 소수점까지 구해라.

```python
weight = 70
height = 1.8
bmi = weight / (height ** 2)
print("당신의 BMI:", bmi)
```

### Q11. 빛의 속도... 아니 시간 변환기
분(minute)을 입력받아 초(second)로 변환해라.

```python
minutes = int(input("몇 분?: "))
seconds = minutes * 60
print(minutes, "분은", seconds, "초다 ㅋㅋㅋ")
```

### Q12. 팩트와 가짜뉴스 구분
어떤 문장을 입력받아 그 안에 "공산주의" 라는 단어가 있으면 "가짜뉴스 검거!" 라고 출력해라.

```python
sentence = input("문장을 써라: ")
if "공산주의" in sentence:
    print("가짜뉴스 검거! ㅋㅋㅋ")
else:
    print("참된 팩트의 문장이구만.")
```
