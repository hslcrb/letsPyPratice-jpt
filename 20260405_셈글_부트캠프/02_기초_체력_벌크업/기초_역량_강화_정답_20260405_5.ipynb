### Q16. 오븐 시계 (BOJ 2525)
[문제 링크](https://www.acmicpc.net/problem/2525)
요리가 끝나는 시각을 정확히 계산해라! ㅋㅋㅋ

```python
h, m = map(int, input().split())
d = int(input())
h += d // 60
m += d % 60
if m >= 60:
    h += 1
    m -= 60
if h >= 24:
    h %= 24
print(h, m)
```

### Q17. 주사위 세개 (BOJ 2480)
[문제 링크](https://www.acmicpc.net/problem/2480)
도박은 나쁘지만 주사위 상금 계산은 코딩 연습에 좋다 ㅋㅋㅋ

```python
a, b, c = map(int, input().split())
if a == b == c:
    print(10000 + a * 1000)
elif a == b or a == c:
    print(1000 + a * 100)
elif b == c:
    print(1000 + b * 100)
else:
    print(max(a, b, c) * 100)
```

### Q19. BMI 계산기 정식 버전
몸무게(kg)와 키(cm)를 입력받아 BMI를 계산하고 판정해라.
BMI = kg / (m * m)

```python
w, h = map(float, input("몸무게(kg)와 키(cm): ").split())
h = h / 100
bmi = w / (h * h)
print("BMI:", round(bmi, 2))
if bmi >= 25:
    print("비만 (관리해라 ㅋㅋㅋ)")
else:
    print("정상 (장 건강 굿!)")
```
