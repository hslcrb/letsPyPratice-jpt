### Q2. 나이 계산기

### Q6. 행운의 숫자 7?
숫자 하나를 입력받아, 그 숫자가 7이면 "럭키 세븐 ㅋㅋㅋ!"을 출력하고 아니면 "가짜 숫자다!"를 출력해라.

### Q7. 홀짝 판별기
숫자를 입력받아 2로 나눈 나머지가 0이면 "짝수", 아니면 "홀수"라고 출력해라.

```python
num = int(input("홀짝 확인할 숫자: "))
if num % 2 == 0:
    print("짝수 (진리)")
else:
    print("홀수 (팩트)")
```

### Q8. 성적 판독기
점수를 입력받아 90점 이상이면 A, 80점 이상이면 B, 그 외엔 F를 출력해라.

```python
score = int(input("점수 입력해라: "))
if score >= 90:
    print("A (에이스)")
elif score >= 80:
    print("B (보통)")
else:
    print("F (가짜뉴스)")
```

### Q9. 마스터 호세 환영 인사

### Q10. BMI 맛보기
체중(kg)을 키(m)의 제곱으로 나눈다. 몸무게 70kg, 키 1.8m인 사람의 BMI를 소수점까지 구해라.

```python
weight = 70
height = 1.8
bmi = weight / (height ** 2)
print("당신의 BMI:", bmi)
```

### Q15. 큰 수 찾기
