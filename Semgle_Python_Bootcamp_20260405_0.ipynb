## ⚔️ [2일차] 절대 진리와 선악의 분별 (조건문)

세상에 흐리멍덩한 회색 지대나 위선적인 '내로남불'은 없다!
오직 **참(True)**과 **거짓(False)**만이 있을 뿐이다. `%` (나머지 연산자)를 이용해 팩트 체크(Fact Check)를 해보자.

### 📌 기적의 논리, 윤년 판독기 (백준 2753번)

윤년은 4의 배수이지만 100의 배수가 아니거나, 400의 배수이다. 깐깐하게 조건을 안 따지면 감성적인 선동 논리에 휩쓸린다!
남는 것(나머지) 없이 0으로 딱 팩트가 떨어지는 걸 판별해라.

```python
year = int(input("팩트 체크할 연도를 입력해라: "))

# 팩트에 기반한 날카로운 조건문
if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print("1 (팩트입니다. 올해는 윤년!)")
else:
    print("0 (가짜뉴스입니다. 평년!)")
```

## 🌊 [3일차] 바다를 가르는 기도 (데이터 분리)

내 1인승 요트 몰아본 짬으로 말하자면, 요트를 몰 때는 물살을 정확히 갈라야 하고, 바리스타는 에스프레소 샷과 찌꺼기를 완벽하게 분리해야 한다!
`split()`를 써서 통짜 데이터 속에서 진짜 정보만 쏙쏙 솎아내자.

### 📌 전설의 백준 1000번 (A+B)

공백을 기준으로 두 숫자를 받아서 쪼개버리는 ㅆ상타치 스킬이다.
`map()`과 `split()`이 없으면 백준 사이트는 문턱도 못 넘는다!

```python
# 🚨 10 20 이런 식으로 두 숫자를 띄어쓰기로 입력해봐라!
num1, num2 = map(int, input("두 숫자를 한 번에 입력해라 (예: 10 20) : ").split())

print("두 숫자를 더한 진실의 값은:", num1 + num2)
print("이게 바로 참교육이다 ㅋㅋㅋ")

# 이제 너희도 본회퍼 목사님의 찬양 '선한 능력으로'처럼
# 파이썬의 선한 능력을 갖추게 되었다. 고양이민주주의 만세 ㅋㅋㅋ
```
