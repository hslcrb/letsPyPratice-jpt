## Part 1: 문자열 기초 정답

```python
# Q1. 문자열 길이
s = input()
print(len(s))
```

```python
# Q2. 아스키 코드
c = input()
print(ord(c))
```

```python
# Q3. 숫자의 합
n = int(input())
s = input()
total = 0
for c in s:
    total += int(c)
print(total)
```

```python
# Q4. 알파벳 찾기
s = input()
for c in 'abcdefghijklmnopqrstuvwxyz':
    print(s.find(c), end=' ')
```

```python
# Q5. 문자열 반복
t = int(input())
for _ in range(t):
    r, s = input().split()
    r = int(r)
    result = ''
    for c in s:
        result += c * r
    print(result)
```

```python
# Q6. 단어 개수
s = input().strip()
words = s.split()
print(len(words))
```

```python
# Q7. 상수
a, b = input().split()
a_rev = int(a[::-1])
b_rev = int(b[::-1])
print(max(a_rev, b_rev))
```

```python
# Q8. 다이얼
s = input()
time = 0
for c in s:
    if c in 'ABC':
        time += 3
    elif c in 'DEF':
        time += 4
    elif c in 'GHI':
        time += 5
    elif c in 'JKL':
        time += 6
    elif c in 'MNO':
        time += 7
    elif c in 'PQRS':
        time += 8
    elif c in 'TUV':
        time += 9
    else:
        time += 10
print(time)
```

```python
# Q9. 크로아티아 알파벳
s = input()
croatia = ['c=', 'c-', 'dz=', 'd-', 'lj', 'nj', 's=', 'z=']
for c in croatia:
    s = s.replace(c, '*')
print(len(s))
```

```python
# Q10. 그룹 단어 체커
n = int(input())
count = 0
for _ in range(n):
    word = input()
    seen = []
    is_group = True
    for i, c in enumerate(word):
        if c in seen and c != word[i-1]:
            is_group = False
            break
        if c not in seen:
            seen.append(c)
    if is_group:
        count += 1
print(count)
```

## Part 2: 리스트 기초 정답

```python
# Q11. 개수 세기
n = int(input())
nums = list(map(int, input().split()))
v = int(input())
print(nums.count(v))
```

```python
# Q12. X보다 작은 수
n, x = map(int, input().split())
nums = list(map(int, input().split()))
for num in nums:
    if num < x:
        print(num, end=' ')
```

```python
# Q13. 최소, 최대
n = int(input())
nums = list(map(int, input().split()))
print(min(nums), max(nums))
```

```python
# Q14. 최댓값
nums = []
for _ in range(9):
    nums.append(int(input()))
max_val = max(nums)
print(max_val)
print(nums.index(max_val) + 1)
```

```python
# Q15. 공 넣기
n, m = map(int, input().split())
baskets = [0] * (n + 1)

for _ in range(m):
    i, j, k = map(int, input().split())
    for idx in range(i, j + 1):
        baskets[idx] = k

for i in range(1, n + 1):
    print(baskets[i], end=' ')
```

```python
# Q16. 공 바꾸기
n, m = map(int, input().split())
baskets = list(range(n + 1))

for _ in range(m):
    i, j = map(int, input().split())
    baskets[i], baskets[j] = baskets[j], baskets[i]

for i in range(1, n + 1):
    print(baskets[i], end=' ')
```

```python
# Q17. 과제 안 낸 사람
all_nums = set(range(1, 31))
submitted = set()
for _ in range(28):
    submitted.add(int(input()))

missing = sorted(all_nums - submitted)
for num in missing:
    print(num)
```

```python
# Q18. 나머지
remainders = set()
for _ in range(10):
    num = int(input())
    remainders.add(num % 42)
print(len(remainders))
```

```python
# Q19. 바구니 뒤집기
n, m = map(int, input().split())
baskets = list(range(n + 1))

for _ in range(m):
    i, j = map(int, input().split())
    baskets[i:j+1] = baskets[i:j+1][::-1]

for i in range(1, n + 1):
    print(baskets[i], end=' ')
```

```python
# Q20. 평균
n = int(input())
scores = list(map(int, input().split()))
m = max(scores)

new_scores = [(s / m) * 100 for s in scores]
print(sum(new_scores) / n)
```

## Part 3: 실전 시나리오 정답

```python
# Q21. 바리스타 메모 파싱
memo = "아메리카노3,에스프레소2,라떼1"
items = memo.split(',')
total = 0
for item in items:
    count = int(item[-1])
    total += count
print(f'총 {total}잔')
```

```python
# Q22. 고양이민주주의 투표 집계
votes = ['찬성', '반대', '찬성', '찬성', '반대', '찬성']
yes = votes.count('찬성')
no = votes.count('반대')

if yes > no:
    print(f'고양이민주주의 통과! (차이: {yes-no}표)')
else:
    print(f'선동 detected... (차이: {no-yes}표)')
```

```python
# Q23. 원전 로그 분석
output_rates = [92, 88, 95, 91, 89, 94, 90, 87, 93, 96]
avg = sum(output_rates) / len(output_rates)

if avg >= 90:
    print('에너지 안보 확보!')
else:
    print('점검 필요')
```

```python
# Q25. 마스터의 비밀 메시지
message = "고양이 민주주의 세션 마스터 호세"
words = message.split()
result = ''
for word in words:
    result += word[0]
print(result)  # 고민세마호
```

```python
# Q26. 커피 원두 재고 관리
stocks = [100, 80, 95, 60, 70]
names = ['아라비카', '로부스타', '콜롬비아', '에티오피아', '케냐']

for i, stock in enumerate(stocks):
    if stock <= 70:
        print(f'{names[i]}: 재고부족({stock}kg)')
```

```python
# Q27. 공산주의 단어 필터링
banned = ['공산주의', '탈원전', '감성팔이', '위선']
text = input('문장: ')

found = False
for word in banned:
    if word in text:
        found = True
        break

print('검열처리' if found else '정상')
```

```python
# Q28. 학생 점수 통계
scores = list(map(int, input('5명 점수: ').split()))
print(f'평균: {sum(scores)/len(scores):.2f}')
print(f'최고: {max(scores)}')
print(f'최저: {min(scores)}')
```

```python
# Q29. 팰린드롬 체커
word = input()
is_palindrome = word == word[::-1]
print('팰린드롬!' if is_palindrome else '아님')
```
